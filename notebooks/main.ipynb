```python
import pandas as pd
from pyproj import Transformer
import folium
import pandas as pd
```

```python
data = pd.read_csv("E:/coding/github/line_between_two_coordinates/data/brasil_city.csv").head(20)

x_origin = data['latitude_origin']
y_origin =  data['longitude_origin']

x_destination = data['latitude_destination']
y_destination = data['longitude_destination']

df = pd.DataFrame({#'name_origin (f)' : ['A', 'A', 'A', 'A', 'A', 'A', 'A'],
                   'x_origin' : x_origin,
                   'y_origin' : y_origin,
                    #'name_destination' : ['Point 1', 'Point 2', 'Point 3', 'Point 4', 'Point 5', 'Point 6', 'Point 7'],
                    'x_destination': x_destination ,
                    'y_destination': y_destination })    
```

```python
latlong_origin = list(zip(df["x_origin"], df["y_origin"]))
latlong_destination = list(zip(df["x_destination"], df["y_destination"]))
m = folium.Map(latlong_origin[0], zoom_start=15)
```

```python
for origin, destination in zip(latlong_origin, latlong_destination):
    folium.CircleMarker([origin[0], origin[1]],
                        radius=15,
                        fill_color="#3db7e4", # divvy color
                       ).add_to(m)

    folium.CircleMarker([destination[0], destination[1]],
                        radius=15,
                        fill_color="red", # divvy color
                       ).add_to(m)

    folium.PolyLine([[origin[0], origin[1]], 
                     [destination[0], destination[1]]]).add_to(m)
m
```
